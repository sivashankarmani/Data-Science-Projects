```python
#Importing Required Lib's
import pandas as pd
import numpy as np
import os
```

```python
#Reading the csv file
df = pd.read_csv("Demand_txn.csv",index_col=0)
```

```python
#Check for NaN values
df.isnull().sum()
```

```python
#Replace NaN values using fillna function
df=df.fillna({'Percent_Quantity_With_Discount':df['Percent_Quantity_With_Discount'].median(),
              'Avg_Discount_Percent_On_Discounted_Items':df['Avg_Discount_Percent_On_Discounted_Items'].median(),
              'Effective_Price':df['Effective_Price'].median(),
              'F_and_B_Inflation_Adjusted_Effective_Price':df['F_and_B_Inflation_Adjusted_Effective_Price'].median(),
              'Month_Of_The_Year_Seasonality':df['Month_Of_The_Year_Seasonality'].median(),
             })
```

```python
df.isnull().sum()
```

```python
#print last 3 observations
df.tail(3)
```

```python
# find the data types of each column - Method 1
df.dtypes
```

```python
# find the data types of each column - Method 2
df.info()
```

```python
#Convert the data type of â€˜Mapped_Item_Codeâ€™ field to object
df = df.astype({"Mapped_Item_Code": str})
```

```python
df.dtypes
```

```python
#Convert the field â€˜Transaction_Dateâ€™ to datetime (Date) column
df['Transaction_Date'] = pd.to_datetime(df['Transaction_Date'])
```

```python
df.dtypes
```

```python
#Extract only those records whose transaction date is greater than or equal to â€˜2017-01-01â€™ 
df1 = df.loc[df['Transaction_Date']>'2017-01-01',:]
```

```python
df1.head()
```

```python
#â€¢	Check the changed data types of each column and also see how many unique sales types are available in â€˜Mapped_Sales_Typeâ€™ column
df1['Mapped_Sales_Type'].unique()
```

```python
#â€¢	Extract only those records whose â€˜Avg_Discount_Percent_On_Discounted_Itemsâ€™ is lesser than 1.0 
df2 = df1.loc[df['Avg_Discount_Percent_On_Discounted_Items']<1.0,:]
```

```python
#and do a sense check if the resulting data frame contains any values in the same column that are greater than or equal to 1.0
df2.loc[df['Avg_Discount_Percent_On_Discounted_Items']>=1.0,:].size
```

```python
#â€¢	Create the below dictionary (dict_agg) and pass it as an argument for the aggregate function in order to roll up the results grouped by following order: 'City', 'Mapped_Sales_Type', 'Mapped_Item_Code', 'Transaction_Date'
dict_aggr = {'Quantity_Sold':'sum', 'Median_Price':'median', 'Effective_Price':'median'}
df1.groupby(['City','Mapped_Sales_Type','Mapped_Item_Code','Transaction_Date']).agg(dict_aggr).reset_index()
```

```python
#â€¢	Print first 5 observations where Effective_Price is not null
df1.loc[df['Effective_Price']!=np.NaN,:].head(5)
```

```python
#â€¢	Print first 10 observations where â€˜Cityâ€™ is â€˜Hyderabadâ€™ AND â€˜Mapped_Sales_Typeâ€™ is â€˜Deliveryâ€™
df1.loc[(df['City']=='Hyderabad')& (df['Mapped_Sales_Type']=='Delivery'),:].head(10)
```

```python
#â€¢	Print first 3 observations where Percent_Quantity_With_Discount is not equal to zero
#df1.loc[(df['Percent_Quantity_With_Discount']!=0)&(df['Percent_Quantity_With_Discount']!=np.NaN),:].head(3)
df1.loc[df['Percent_Quantity_With_Discount']!=0,:].head(3)
```

```python
#â€¢	Create a new column â€˜Rangeâ€™ and populate it as 1 if Effective_Price is greater than 500, else 0.5
df1['Range'] = np.where(df1['Effective_Price']>500, 1 ,0.5)
```

```python
df1.columns
```

```python
df1.head(10)
```

```python
#â€¢	Split the dataset into X and Y based on below list:
x = df1[['Mapped_Item_Code', 'Median_Price', 'Average_Tax_Percent','Effective_Price','Percent_Quantity_With_Discount'
         ,'Avg_Discount_Percent_On_Discounted_Items','Car_sales','F_and_B_Inflation',
         'F_and_B_Inflation_Adjusted_Effective_Price','F_and_B_Inflation_Adjusted_Median_Price']]

y = df1['Quantity_Sold']
```

```python
#Check for NAN values
x.isnull().sum()
```

```python
x.dtypes
```

```python
#Split X train and Y train , X Test and Y Test
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3, random_state=100)
```

```python
#Import Linear Regression
from sklearn.linear_model import LinearRegression
```

```python
#Assign Object
lr = LinearRegression()
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
#Fit the data to algorithm
lr.fit(x_train,y_train)
```

```python
#Predict Train prediction
y_train_predict = lr.predict(x_train)
```

```python
from sklearn import metrics
metrics.mean_squared_error(y_train,y_train_predict)
```

```python
#Predict Test prediction
y_predict = lr.predict(x_test)
```

```python
metrics.mean_squared_error(y_test,y_predict)
```

```python
#R-Sqr of the model fit - for training
print("R-Srq value of this fit: ",metrics.r2_score(y_train,y_train_predict))
```

```python
#R-Sqr of the model fit - for test
print("R-Srq value of this fit: ",metrics.r2_score(y_test,y_predict))
```
